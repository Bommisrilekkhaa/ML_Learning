```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
col = ["mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness"]
from sklearn.preprocessing import MinMaxScaler
st = MinMaxScaler()
data[col] = st.fit_transform(data[col])
data
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import lightgbm as lgb
model = lgb.LGBMClassifier( n_estimators = 20,learning_rate = 0.7,reg_lambda=0.8)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
acc
```
