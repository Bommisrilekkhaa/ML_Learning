#Support vector machine

kernel trick

1. binary classification

2. linearly separable --> linear classification

3. decision boundary

4. maximum margin hyperplane

5. find optimal line (hyperplane)

#vector

1. direction --> magnitude

#margin

--> margin should be maximal for both classes

--> margin larger,maximum

--> r^2 - line

--> r^3 - plane

#converting 2d to 3d:

1. for classifying non linear data

2. converting 2d to 3d --> kernel trick

3. non linear data --> separated as layeres --> can be separated by hyperplane easily

#formulas:

1. linear kernel --> k(x,y) = x*y

2. polynomial --> k(x,y) = (x*y + 1)^d

3. RBF --> k(x,y) = e^(-gamma(mod(x-y)^2))



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data[data.Embarked.isnull()]
```

```python
data.drop(['Cabin'],axis=1,inplace=True)
```

```python
data['Age'] = data["Age"].fillna(data.Age.median())
data["Embarked"] = data["Embarked"].fillna("C")
data.Sex = data.Sex.replace(['male','female'],[0,1])
data.Embarked = data.Embarked.replace(['S','C',"Q"],[0,1,2])
```

```python
data.dropna(inplace=True)
```

```python
data_test = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
data_test.head()
```

```python
data_test['Age'] = data_test["Age"].fillna(data_test.Age.median())
data_test["Embarked"] = data_test["Embarked"].fillna("C")
data_test.Sex = data_test.Sex.replace(['male','female'],[0,1])
data_test.Embarked = data_test.Embarked.replace(['S','C',"Q"],[0,1,2])
```

```python
data_test.drop(['Cabin'],axis=1,inplace=True)
```

```python
data_test.dropna(inplace=True)
```

```python
features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
```

```python
#gamma #SVC - support vector classifier
c = 0.7
clf = svm.SVC(kernel="linear",C=c).fit(data[features],data["Survived"])
```

```python
pred = clf.predict(data_test[features])
```
